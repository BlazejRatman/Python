### TIC TAC TOE

Display current view of board using 'o', 'x', '-' after all movements

Input 'x' and 'o' are put alternatively according to the number of character position from 1 to 9 as below:
```
123
456
789
```
Finish the game if someone win or board will be full

````python
board = ["-", "-", "-", "-", "-", "-", "-", "-", "-"]


print("TIC TAC TOE")
def show_board():
    print(board[0] + board[1] + board[2])
    print(board[3] + board[4] + board[5])
    print(board[6] + board[7] + board[8])

def game():
    show_board()
    
    while victory() is False:
        if board.count("-") == 0:
            print("Game over. There are no free spaces!")
            break            
        move_x()
        victory()
        if victory() is True:
            print("The winner is X")
            break
        if board.count("-") == 0:
            print("Game over. Tie!")
            break 
        move_o()
        victory()
        if victory() is True:
            print("The winner is O")
            break 

def move_x():
    move_x = input("Input 'x':")

    valid = False
    while not valid:
        
        while move_x not in ["1", "2", "3", "4", "5", "6", "7", "8", "9"]:
            move_x = input("Invalid input")
            print("Invalid input. Please choose position for 'x'")
        print(f"Input 'x': {move_x}")
        move_x = int(move_x) - 1
        if board[move_x] == "-":
            board[move_x] = "x"
            valid = True
            show_board()
        else:
            print("This place is occupied. Please select another one!")

def move_o():
    move_o = input("Input 'o':")

    valid = False
    while not valid:
    
        while move_o not in ["1", "2", "3", "4", "5", "6", "7", "8", "9"]:
            move_o = input("Invalid input")
            print("Invalid input. Please choose position for 'o' ")
        print(f"Input 'o': {move_o}")
        move_o = int(move_o) - 1
        if board[move_o] == "-":
            board[move_o] = "o"
            valid = True
            show_board()
        else:
            print("This place is occupied. Please select another one!")

        

def victory():
    if board[0] == board[1] == board[2] == "x":
        return True
    elif board[3] == board[4] == board[5] == "x":
        return True
    elif board[6] == board[7] == board[8] == "x":
        return True
    elif board[0] == board[3] == board[6] == "x":
        return True    
    elif board[1] == board[4] == board[7] == "x":
        return True
    elif board[2] == board[5] == board[8] == "x":
        return True   
    elif board[0] == board[4] == board[8] == "x":
        return True
    elif board[2] == board[4] == board[6] == "x":
        return True
    elif board[0] == board[1] == board[2] == "o":
        return True
    elif board[3] == board[4] == board[5] == "o":
        return True
    elif board[6] == board[7] == board[8] == "o":
        return True
    elif board[0] == board[3] == board[6] == "o":
        return True    
    elif board[1] == board[4] == board[7] == "o":
        return True
    elif board[2] == board[5] == board[8] == "o":
        return True   
    elif board[0] == board[4] == board[8] == "o":
        return True
    elif board[2] == board[4] == board[6] == "o":
        return True
    else:
        return False

game()
````
